```python
class MetodoPago:
    def __init__(self,monto):
        self.monto=monto
    def procesar_pago(self):
        return f"usted pago esta cantidad {self.monto}"
        
class PagoEfectivo(MetodoPago):
     def __init__(self,monto,tipo,vuelta,montop):
      
        super().__init__(monto)
        self.tipo=tipo
        self.vuelta=vuelta
        self.montop=montop
     def procesar_pago(self):
         self.montop=int(input(f"ingresa la cantidad que desea pagar en {self.tipo} recuerde que la factura es de {self.monto}"))
         d=self.montop-self.monto
         if d>0:
             self.vuelta=d
         else:
             print("no tiene devuelta")
         print(f"el pago se realizo correctamente\n"
         f"------------------------------------------------------------\n"
         f"usted pago esta cantidad {self.montop} a una factura de {self.monto} en {self.tipo} y obtuvo una vuelta de dinero de {self.vuelta} " )
         
class PagoTransferencia(MetodoPago):
     def __init__(self,monto,tipo,cuenta,montop):
      
        super().__init__(monto)
        self.tipo=tipo
        self.cuenta=cuenta
        self.montop=montop
     def procesar_pago(self):
         self.montop=int(input(f"ingresa la cantidad que desea pagar en {self.tipo} recuerde que la factura es de {self.monto}" ))
         self.cuenta=int(input(f"ingrese el numero de cuenta al que desea transferir {self.monto} "))
         print (f"usted pago esta cantidad {self.montop} a una factura de {self.monto} por {self.tipo} a este numero de cuenta {self.cuenta}" )


class PagoTarjeta(MetodoPago):
     def __init__(self,monto,tipo,disponible,montop,clave):
      
        super().__init__(monto)
        self.tipo=tipo
        self.disponible=disponible
        self.montop=montop
        self.clave=clave
     def procesar_pago(self):
         while True:
            p=input("ingrese la clave de la tarjeta (pista -fo**-)")
            if p== self.clave: 
                self.montop=int(input(f"ingresa la cantidad que desea pagar en {self.tipo} recuerde que tiene {self.disponible} de saldo y \n \
                     la factura es de {self.monto}"))
         
                if self.montop<= self.disponible:
                    print (f"usted pago esta cantidad {self.monto} a una factura de {self.monto} por {self.tipo}") 
                    break
                else:
                    print (f"usted notiene saldo disponible en la {self.tipo}" )
                    break
            
            else:
                f"la contraseña que ingresaste{p} no es correcta intenta de nuevo"

pago1=PagoEfectivo(100000,"efectivo",0,0)
pago2=PagoTransferencia(100000,"transferencia",0,0)
pago3=PagoTarjeta(100000,"tarjeta",120000,0,"fofo")
pagos=[]
pagos.append(pago1)
pagos.append(pago2)
pagos.append(pago3)
for i in pagos:
    i.procesar_pago()
```
