```python
class personaje:
    def __init__(self,cantidad_de_vida):
        self.cantidad_de_vida=cantidad_de_vida
        
    def atacar(self):
        return f"esta unidad ataca y hace daño"

class mele(personaje):
    def __init__(self,cantidad_de_vida,daño_espada):
        super().__init__(cantidad_de_vida)
        self.daño_espada=daño_espada

    def atacar(self):
        print(f"esta unidad  mele ataca y hace daño de {self.daño_espada} disminuyendo la vida del oponente") 
 
class mago(personaje):
    def __init__(self,cantidad_de_vida,daño_magia):
        super().__init__(cantidad_de_vida)
        self.daño_magia=daño_magia

    def atacar(self):
        print(f"esta unidad mango ataca y hace daño de {self.daño_magia} disminuyendo la vida del oponente")    
        
class archier(personaje):
    def __init__(self,cantidad_de_vida,daño_flechas):
        super().__init__(cantidad_de_vida)
        self.daño_flechas=daño_flechas

    def atacar(self):
        return f"esta unidad arquero ataca y hace daño de {self.daño_flechas} disminuyendo la vida del oponente"
        

m=mele(120,10)
mag=mago(80,19)
a=archier(100,13)

m.atacar()
mag.atacar()
print(f"{a.atacar()}")
 
```
