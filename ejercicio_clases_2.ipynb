Clase base llamada reserva. Con un atributo llamado Cantidad_de_Pasajeros. Metodo: Confirmar_Transporte. 

2)  2 Clases hijas. Reserva Lancha, y Reserva Bus. Cada una con un limite de pasajeros. Cada una va a confirmar Transporte segun el limite de pasajeros.

```python
class Reserva:
    def __init__(self,Cantidad_de_Pasajeros):
        self.Cantidad_de_pasajeros = Cantidad_de_Pasajeros
    
    def Confirmar_Transporte(self):
        return 0
    
class Reserva_Lancha(Reserva):
    def Confirmar_Transporte(self,Cantidad_de_pasajeros):
        super().__init__(Cantidad_de_pasajeros)
        self.limite = 10
        if self.Cantidad_de_pasajeros >self.limite:
            return print(f"No se pueden reservar mas de {self.limite} pasajeros")
        else:
            return print(f"Pasajeros en lancha {self.Cantidad_de_pasajeros}, listo para salir")

class Reserva_Bus(Reserva):
    def Confirmar_Transporte(self,Cantidad_de_pasajeros):
        super().__init__(Cantidad_de_pasajeros)
        self.limite = 15
        if self.Cantidad_de_pasajeros >self.limite:
            return print(f"No se pueden reservar mas de {self.limite} pasajeros en bus")
        else:
            return print(f"Pasajeros en lancha {self.Cantidad_de_pasajeros}, listo para salir")
pasajeros = int(input("Ingresa la cantidad de pasajeros a reservar"))
opcion= int(input(f"Ingresa (1) para reservar en Lancha, (2) Bus, (3) salir"))

if opcion ==1:
    reserva = Reserva_Lancha(pasajeros)
    reserva = reserva.Confirmar_Transporte(pasajeros)

elif opcion ==2:
    reserva = Reserva_Bus(pasajeros)
    reserva = reserva.Confirmar_Transporte(pasajeros)

elif opcion ==3:
    print("Gracias por usar el sistema de reservas ANDRES S.A.S")
else:
    print(f"Ingresa un valor correcto, valor ingresado {(opcion)}")
```
