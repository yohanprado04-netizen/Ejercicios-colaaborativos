```python
class Documento:
    def __init__(self, titulo, autor, contenido):
        self.titulo = titulo
        self.autor = autor
        self.contenido = contenido

    def mostrar_informacion(self):
        print(f"Título: {self.titulo}")
        print(f"Autor: {self.autor}")
        print(f"Contenido: {self.contenido}")
        
class DocumentoPDF(Documento):
    def __init__(self, titulo, autor, contenido, numero_paginas):
        super().__init__(titulo, autor, contenido)
        self.numero_paginas = numero_paginas

    def mostrar_informacion(self):
        super().mostrar_informacion()
        print(f"Número de páginas: {self.numero_paginas}")

class DocumentoWord(Documento):
    def __init__(self, titulo, autor, contenido, numero_palabras):
        super().__init__(titulo, autor, contenido)
        self.numero_palabras = numero_palabras

    def mostrar_informacion(self):
        super().mostrar_informacion()
        print(f"Número de palabras: {self.numero_palabras}")

class DocumentoExcel(Documento):
    def __init__(self, titulo, autor, contenido, numero_hojas):
        super().__init__(titulo, autor, contenido)
        self.numero_hojas = numero_hojas

    def mostrar_informacion(self):
        super().mostrar_informacion()
        print(f"Número de hojas: {self.numero_hojas}")
```
